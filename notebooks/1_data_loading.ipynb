```python
import pandas as pd

# Update with the absolute or relative file path
data_path = 'C:/Users/user/Desktop/DA/data/features.csv'

# Load the dataset
df = pd.read_csv(data_path)

# Display the first few rows
print("Dataset Preview:")
print(df.head())

# Overview of the dataset
print("\nDataset Info:")
print(df.info())

# Check for missing values
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

# Fill missing values
df['CPI'] = df['CPI'].fillna(df['CPI'].mean())  # Fill CPI missing values with the mean
df['Unemployment'] = df['Unemployment'].fillna(df['Unemployment'].mean())  # Fill Unemployment missing values with the mean

# Fill markdown columns with 0
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
for col in markdown_cols:
    df[col] = df[col].fillna(0)

# Check for remaining missing values
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# Statistical summary of numerical columns
print("\nStatistical Summary:")
print(df.describe())

# Validate data types
print("\nData Types:")
print(df.dtypes)
```
